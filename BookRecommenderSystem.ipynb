## Collaborative Based Filtering Algorithms

```python
ratings_with_name
```

```python
ratings_with_name.groupby('User ID').count()['Ratings']
```

```python
x = ratings_with_name.groupby('User ID').count()['Ratings'] >= 5
x[x].index
```

```python
potential_users = x[x].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User ID'].isin(potential_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Ratings'] >=5
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User ID',values='Ratings')
```

```python
pt.shape
```

```python
#pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#now we have our pivot table , we only have to calculate the distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape #shape shows that each book has similarity with every book
```

```python
similarity_scores
```

```python
similarity_scores[0] #similarity of book 1 with all other books of pivot table
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse = True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image URL'].values))
        
        data.append(item)
    
    return data
```
